```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import seaborn as sn
sn.set_theme(style="whitegrid")
```

```python
def plot_setup(ax1, ax2 ,title, x, time, all_visited, label=True):
    ax1.bar(x,all_visited, color='b', label="Search space")
    
    
    if label:
        leg1= ax1.set_ylabel("Number of visited nodes")
        ax1.set_yscale("log")
        ax1.yaxis.label.set_color('blue')
        #ax1.spines['left'].set_color('blue')
        ax1.tick_params(axis='y', colors='blue')

    leg2=ax2.plot(x,time, label="Time", c='r')
    if label:
        ax2.set_ylabel("Runtime in seconds")
        ax2.set_yscale("log")
        ax2.yaxis.label.set_color('red')
        #ax2.spines['right'].set_color('red')
        ax2.tick_params(axis='y', colors='red')


    #ax1.legend(loc=0)
    #ax2.legend(loc=1)
    ax1.set_title(title)
```

```python
x_proof = ["4x9", "4x10", "4x11", "4x12", "4x13", "4x14"]
all_visited_proof = [7115, 52138, 191612, 1341134, 14341212, 261965503]
time_proof = [0.055, 0.34, 1.59, 10.07, 172, 46*60+20]
```

```python
plt.figure(figsize=(10,8))
ax1 = plt.gca()
ax12 = ax1.twinx()
plot_setup(ax1, ax12, "Disproof setup", x_proof, time_proof, all_visited_proof)
plt.show()
```

```python
x_disproof = ["4x7", "4x8", "4x9", "4x10", "4x11"]
all_visited_disproof = [89217,   578881,  9606929, 62171146, 639454658]
time_disproof = [0.38, 3.13, 78.21, 915.85, 14143.59]
```

```python
plt.figure(figsize=(10,8))
ax1 = plt.gca()
ax12 = ax1.twinx()
plot_setup(ax1, ax12, "Disproof setup", x_disproof, time_disproof, all_visited_disproof)
plt.show()
```

```python
fig,(ax1,ax2) = plt.subplots(1,2, figsize=(12,8), sharey=True)

ax1.get_shared_y_axes().join(ax1, ax2)

ax12 = ax1.twinx()
ax22 = ax2.twinx()


ax12.get_shared_y_axes().join(ax12, ax22)

ax_share = plot_setup(ax1, ax12, "Proof setup", x_proof, time_proof, all_visited_proof, False)

plot_setup(ax2, ax22, "Disproof setup", x_disproof, time_disproof, all_visited_disproof, False)

if True:
    ax1.set_ylabel("Number of visited nodes")
    ax1.set_yscale("log")
    ax1.yaxis.label.set_color('blue')
    #ax1.spines['left'].set_color('blue')
    ax1.tick_params(axis='y', colors='blue')

    ax22.set_ylabel("Runtime in seconds")
    ax22.set_yscale("log")
    ax22.yaxis.label.set_color('red')
    #ax2.spines['right'].set_color('red')
    ax22.tick_params(axis='y', colors='red')
    
    ax12.get_yaxis().set_visible(False)
    ax2.get_yaxis().set_visible(False)
    
    ax1.legend(loc=2)
    ax22.legend(loc=1)

plt.show()
```
