```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [i for i in range(9,15)]
x = ["4x9", "4x10", "4x11", "4x12", "4x13", "4x14"]
all_visited = [7115, 52138, 191612, 1341134, 14341212, 261965503]
time = [0.055, 0.34, 1.59, 10.07, 172, 46*60+20]
```

```python
plt.figure(figsize=(10,8))
ax1 = plt.gca()
ax2 = ax1.twinx()

ax1.plot(x,all_visited, c='b', label="Size")
leg1= ax1.set_ylabel("Number of visited nodes")
ax1.set_yscale("log")
ax1.yaxis.label.set_color('blue')
#ax1.spines['left'].set_color('blue')
ax1.tick_params(axis='y', colors='blue')

leg2=ax2.plot(x,time, label="Time", c='r')
ax2.set_ylabel("Runtime in seconds")
ax2.set_yscale("log")
ax2.yaxis.label.set_color('red')
#ax2.spines['right'].set_color('red')
ax2.tick_params(axis='y', colors='red')

ax1.legend(loc=0)
ax2.legend(loc=1)
plt.title("Proof progress")
#plt.legend()
```
